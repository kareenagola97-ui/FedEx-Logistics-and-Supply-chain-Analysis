Import  Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

Data Loading

```python
data=pd.read_csv("/content/SCMS_Delivery_History_Dataset (1).csv")
```

Data Cleaning

```python
data.isnull().sum()
```

```python
data['Shipment Mode']=data['Shipment Mode'].fillna(0)
data['Dosage']=data['Dosage'].fillna('Unknown')
data['Line Item Insurance (USD)']=data['Line Item Insurance (USD)'].fillna(0)
```

```python
data.duplicated().sum()
```

```python
data_cols=['PQ First Sent to Client Date', 'PO Sent to Vendor Date',
       'Scheduled Delivery Date', 'Delivered to Client Date',
       'Delivery Recorded Date']

for col in data_cols:
      data[col] = pd.to_datetime(data[col],errors='coerce')
```

```python
data['Weight (Kilograms)']=pd.to_numeric(data['Weight (Kilograms)'],errors='coerce')
data['Weight (Kilograms)']=data['Weight (Kilograms)'].fillna(data['Weight (Kilograms)'].median())
data['Freight Cost (USD)']=pd.to_numeric(data['Freight Cost (USD)'],errors='coerce')
data['Freight Cost (USD)']=data['Freight Cost (USD)'].fillna(data['Freight Cost (USD)'].median())
data['PQ First Sent to Client Date']=data['PQ First Sent to Client Date'].fillna(data['PQ First Sent to Client Date'].median())
data['PO Sent to Vendor Date']=data['PO Sent to Vendor Date'].fillna(data['PO Sent to Vendor Date'].median())


```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
data.describe(include='object')
```

EDA

```python
top_10_country_shipment=data['Country'].value_counts().head(10).reset_index()
top_shipment_mode=data['Shipment Mode'].value_counts().head(10).reset_index()
top_vendor=data['Vendor'].value_counts().head(10).reset_index()
top_manufacturing_site=data['Manufacturing Site'].value_counts().head(10).reset_index()
country_genarte_highest_shipment=data.groupby('Country')['Line Item Value'].sum().sort_values(ascending=False).head(10).reset_index()
vendor_performance=data.groupby('Vendor')['Line Item Value'].sum().sort_values(ascending=False).head(10).reset_index()
country_freight_cost=data.groupby('Country')['Freight Cost (USD)'].sum().sort_values(ascending=False).head(10).reset_index()
product_demand=data.groupby('Product Group')['Line Item Quantity'].sum().sort_values(ascending=False).head(10).reset_index()
cost_freight=data.groupby('Shipment Mode')['Freight Cost (USD)'].sum().sort_values(ascending=False).head(10).reset_index()
manufacturing_site=data.groupby('Manufacturing Site')['Line Item Value'].sum().sort_values(ascending=False).head(10).reset_index()
highest_rev_product_brand=data.groupby(['Product Group','Brand'])['Line Item Value'].sum().sort_values(ascending=False).head(10).reset_index()
brand_quantiy=data.groupby('Brand')['Line Item Quantity'].sum().sort_values(ascending=False).head(10).reset_index()
average_weight=data['Weight (Kilograms)'].mean()
freight_avg=data['Freight Cost (USD)'].mean()
print(top_10_country_shipment)
print(top_shipment_mode)
print(top_vendor)
print(top_manufacturing_site)
print(country_genarte_highest_shipment)
print(vendor_performance)
print(country_freight_cost)
print(product_demand)
print(cost_freight)
print(manufacturing_site)
print(highest_rev_product_brand)
print(brand_quantiy)
print(average_weight)
print(freight_avg)


```

**Visualizations**

Total Shipment Mode Count

```python
data['Shipment Mode'].value_counts().plot(kind='pie',autopct='%1.1f%%',title='Shipment Mode',explode=(0.1,0,0,0,0),shadow=True)
plt.show()
```

**Insight**

Air transport is the most frequently used shipment mode,also show the importance of speed and timely delivery in the supply chain.

Top 10 Country and Top Vendors,Brands



```python
top_10_country_shipment=data['Country'].value_counts().head(10).reset_index()
top_vendor=data['Vendor'].value_counts().sort_values(ascending=False).head(10).reset_index()
top_brand=data['Brand'].value_counts().head(10).reset_index()
fig=make_subplots(rows=1,cols=2,subplot_titles=('Top 10 Country','Top Vendor','Top Brand'))
fig.add_trace(go.Bar(x=top_10_country_shipment['Country'],y=top_10_country_shipment['count']),row=1,col=1)
fig.add_trace(go.Bar(x=top_brand['Brand'],y=top_brand['count']),row=1,col=2)
fig.update_layout(height=500,width=1100,title_text='Top 10 Country,Top Vendor,Top Brand')
fig.show()
fig=px.bar(top_vendor,x='count',y='Vendor',title='Top 10 Vendor',orientation='h',color='count')
fig.update_layout(height=500,width=1100)
fig.show()


```

**Insight**

This chart show the top country has more shipment in south africa and SCMS from RDC and has more shipment vendor is generic.

Freight Cost and Weight Distribution

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
fig=plt.hist(np.log1p(data['Freight Cost (USD)']))
plt.title('Freight Cost (USD)')
plt.xlabel('Freight Cost (USD)')
plt.ylabel('Frequency')
plt.subplot(1,2,2)
fig=plt.hist(np.log1p(data['Weight (Kilograms)']),color='red')
plt.title('Weight (Kilograms)')
plt.xlabel('Weight (Kilograms)')
plt.ylabel('Frequency')
plt.show()
```

**Insight**

This chart show the frequency of freight cost and also weight distribution with low shipment and few shipment .

Yearly Revenue Trend

```python
data['Delivered Year']=data['Delivered to Client Date'].dt.year
Yearly_revenue=data.groupby('Delivered Year')['Line Item Value'].sum().reset_index()
fig=px.line(Yearly_revenue,x='Delivered Year',y='Line Item Value',title='Yearly Revenue Trend',markers="-")
fig.update_xaxes(title_text='Year')
fig.update_yaxes(title_text='Revenue')
fig.update_layout(height=500,width=1100)
fig.show()
```

**Insight**

Revenue shows variation across the year ,the 2014 more revenue.

Shipments Trend by Month

```python
data['Delivered Month']=data['Delivered to Client Date'].dt.month
monthly_shipment=data.groupby('Delivered Month').size()
fig=px.line(monthly_shipment,x=monthly_shipment.index,y=monthly_shipment.values,title='Shipments Trend by Month',markers="-")
fig.update_xaxes(title_text='Month')
fig.update_yaxes(title_text='Shipments')
fig.update_layout(height=500,width=1100)
fig.show()
```

**Insight**

Monthly shipment that also show the variation in month ,this chart has the 8 month of shipment more.

Revenue by Country

```python
country_revenue=data.groupby('Country')['Line Item Value'].sum().sort_values(ascending=False).reset_index()
fig=px.bar(country_revenue,x='Country',y='Line Item Value',title='Revenue by Country',color='Line Item Value')
fig.update_xaxes(title_text='Country')
fig.update_yaxes(title_text='Revenue')
fig.update_layout(height=500,width=1100)
fig.show()
```

**Insight**

They show the variation by country,nigeria has more revenue the other country and showing the variations.

Weight vs Freight Cost

```python
fig=px.scatter(data,x='Weight (Kilograms)',y='Freight Cost (USD)',title='Weight vs Freight Cost',color='Shipment Mode')
fig.update_layout(height=500,width=1100)
fig.show()
```

**Insight**

Freight cost generally increase  with shipment weight,indicating a positive relationship between transportation cost and cargo weight.

Correlation Heatmap

```python
corr=data[['Weight (Kilograms)','Freight Cost (USD)','Line Item Value','Line Item Quantity']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
```

**Insight**

Freight cost has a positive correlation with shipment weight and quantity,indicating that larger shipments generally require higher transportation spending.
